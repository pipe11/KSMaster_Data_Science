### Machine learning
Técnicas de computación que usa estadística para habilitar los sistemas de computación la habilidad de "aprender" con datos, modelos.

Automatizar toda la cadena de producción. Desventaja: Peor calidad delproducto. AKA Toyota. Lo ideal, que tareas automatizar y cuales no.

Cognitive tasks, automated with machine learning. Automatizar tareas cognitivas sencillas. Tipos de aprendizaje (99,9% generalización/relación y comparación):

   * Generalización y relación: Circulo relleno V, cuadrado vacío X, triángulo relleno X, circulo vacío V. ¿Qué entra? Según características, target y relación de lo que ha entrado, llegamos a la conclusión de que lo que entra es el círculo. Y generaliza a que todas las figuras con forma de círculo, entran. F(características) = target.
   
   * Comparación: Extraemos características (círtricos, colores, formas, acidez, origen, hoja). Distancia entre las características (d1, d2). Grupo segmento.
   
   * Refuerzo: Se basa por la recompensa. Va a intentar varias acciones, hasta que consiga la recompensa.

### ¿Qué tipos de tareas cognitivas?:

   * Supervised learning: En el supervisado tienes etiquetas. Problemas: Hay que etiquetar continuamente la info. Aprende por elementos pasados:
       * Regression: Predice el número.
       * Clasificación de etiquetas, predice etiquetas.
   
   * Semi-supervisado: Hay un etiquetado parcial y el resto por comparación. Aprende por comparación:
       * Clustering: Encuentra los elementos distintos.
       * Reducción de dimensionalidad: Explica los elementos sin tantos atributos
   
   * Unsupervised learning: No etiqueto los elementos, lo hago por comparaciones.

### Machine Learning example: Prediciendo precios de casas

Problema: Precio de la casa según las características.

Solución: De precios de casa extraemos las características. Extraemos datos. Categorizamos las características. Cada línea representa una casa. Entre una etiqueta/target, establecer una correlación.

Correlación es un término de estadística. Está muy ligada a las regresiones lineales. No es el único criterio para saber si una variable es buena o no.

¿Que características o variables habrá que coger?
El saber hacer de Data Science es necesario un knowledge expertise bueno, para elegir bien una buena relación entre target y cracterísticas. Es necesario elegir bien las características que mejoren el modelo de predicción. Añadir cuantas más características que mejoren el modelo.

#### Modelo: Simplificación de la realidad.

   * Example of model 1: Regresión lineal. Intenta enontrar una función lineal. Necesitamos encontrar a,b,c y d que tengan una mejor relación.
   
   * Example of model 2: Decision tree. Un arbol de decisión trata de encontrar un path para explicar el target. Divdimos casos entre varias características. Necesitamos encontrar los mejores splits para nuestros modelos.
   
   * Randomon Forest, K-neighbors, SVM, Neural Networks, Deep Learning y la lista de modelos crece.
   
Deep Learning, como lo hace: Los pcs son gilipollas, no tienen esa capacidad de abstracción para la regresión lineal. Quiero guardar el millón de puntos. No simplificamos con matemáticas, nos quedamos con todos los datos, incluso aquellos que aparentemente no tienen relación.

### Parámetros del modelo

El Data Scientist trabaja para encontrar los mejores parámetros del modelo.

Parámetros, que peso de decisión le doy a una características. Lo hace automáticamente el algoritmo, que es quien encuentra los parámetros y le da el peso. Con las estadística podemos modelizar la incertidumbre.

¿Cómo encuentras los mejores parámetros?

Training: Cada modelo tiene un particular way para encontrar los mejores parámetros.

Hay muchas métricas que pueden ser elegidas para una tarea:

   * Bias: Media de los errores.
   * MAE: MEdia del valor absoluto de los errores.
   * RMSE:  La media del cuadrado de los errores.
   * Hay más...
   
Elegimos los parametros del modelo que nos dan la mejor métrica. Elegiremos el modelo que tenga menor media de errores. El negocio o la necesidad de negocio, estableceremos las métricas que más se ajusten a nuestras necesidades para elegir el modelo.

FE: Cogemos las características disponibles en tabla y dividimos en 2 el set:

   * 1 set para calcular los parámetros: Training set. (Es como el 80-90%)
   * 1 set para calcular las métricas: Test set.
