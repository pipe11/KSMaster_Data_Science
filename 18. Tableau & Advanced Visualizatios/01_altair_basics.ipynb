## Visualizations with Google trends.csv

```
trends = pd.read_csv("https://raw.githubusercontent.com/rezpe/datos_viz/master/google_trends.csv")
trends.sample(10)
```

```
trends.columns
```

```
comp_trends = alt.Chart(trends).mark_bar().encode(
    x = 'search_term',
    y = 'mean(value)',
    color = 'search_term'
).properties(
    title = 'Composición de tendencias',
    width = 200
).interactive()

comp_trends
```

```
line_trends = alt.Chart(trends).mark_line().encode(
    x = 'yearmonth(date):T',
    y = 'mean(value)',
    color = 'search_term'
).properties(
    title = 'Gráfico histórico de tendencia de búsquedas',
    width = 500
).interactive()

line_trends
```

```
select_st = alt.selection(type="interval",encodings=["x"])

comp_trends = alt.Chart(trends).mark_bar().encode(
    x='search_term',
    y='mean(value)',
    color=alt.condition(select_st, 'search_term', alt.value('lightgray')),
    tooltip="search_term"
).properties(
    width=100,
    selection=select_st
)

trends_line = alt.Chart(trends).mark_line().encode(
    x="yearmonth(date):T",
    y='mean(value)',
    color="search_term"
).properties(
    width=600
).transform_filter(
    select_st
)

trends_dash = (comp_trends|trends_line).properties(
    title="Búsquedas de Google de términos de data science"
)

trends_dash
```

```
trends_dash.save("trends_dash.html")
```

### Ejercicio: Select y filtrar ambos histogramas hacia el scatter

```
select_points = alt.selection(type="interval",encodings=["x","y"])

hist_brain = alt.Chart(brain).mark_bar().encode(
    x=alt.X('Brain Weight',bin=alt.Bin(maxbins=10)),
    y="count()"
).properties(
    title="Distribución del peso de los cerebros",
    width=400
).transform_filter(
    select_points
)

hist_body = alt.Chart(brain).mark_bar().encode(
    x=alt.X('Body Weight',bin=alt.Bin(maxbins=10)),
    y="count()"
).properties(
    title="Distribución del peso de los cuerpos",
    width=400
).transform_filter(
    select_points
)

scatter_brain_body = alt.Chart(brain).mark_point().encode(
    x='Brain Weight',
    y='Body Weight'
).properties(
    title="Distribución del peso de los cerebros",
    width=800,
    selection=select_points
)

brain_filter_dash = scatter_brain_body & (hist_body | hist_brain)
brain_filter_dash
```

### Ejercicio: Gráfico de barras comparandos la temperatura media por países

```
# Nos va a saltar error porque el csv es muy grandes. Para eso usamos:
alt.data_transformers.disable_max_rows()
```

```
temp = weather_df.groupby("country").mean().reset_index()
hist_temp_countries = alt.Chart(temp).mark_bar().encode(
    x = alt.X('country', 
              sort = alt.Sort(field = 'temp', 
                                         op = 'mean', 
                                         order = 'descending'),
              axis = None
              ),
    y="mean(temp)",
    tooltip = 'country'
).properties(
    title="Temperatura media de los países",
    width = 800
).interactive()

hist_temp_countries
```

### Ejercicio: Gráfico de barras comparandos la temperatura media por países

```
select_ctry = alt.selection(type="single",encodings=["x"])

temp = weather_df.groupby(["city","country"]).mean().reset_index()

ranking_city = alt.Chart(temp).mark_bar().encode(
    x=alt.X("city",
            sort=alt.Sort(field="temp",
                          op="mean",
                          order="descending"),
            
            ),
    y="mean(temp)",
    tooltip="city"
).properties(
    width=800,
    title="Ranking de ciudades por temperatura media"
).transform_filter(
    select_ctry
)

ranking_country = alt.Chart(temp).mark_bar().encode(
    x=alt.X("country",
            sort=alt.Sort(field="temp",
                          op="mean",
                          order="descending"),
            axis=None
            ),
    y="mean(temp)",
    tooltip="country"
).properties(
    width=800,
    title="Ranking de ciudades por temperatura media",
    selection=select_ctry
)

dash_rankings = ranking_country&ranking_city

dash_rankings
```

```
countries = alt.topo_feature("https://cdn.jsdelivr.net/npm/world-atlas@2/countries-110m.json",
                             feature="countries")

background = alt.Chart(countries).mark_geoshape(
    fill="lightgrey",
    stroke="blue"
)

temp = weather_df.groupby("city").mean().reset_index()
cities = alt.Chart(temp).mark_point().encode(
    latitude="lat",
    longitude="lon",
    size=alt.value(1),
    color=alt.Color("temp",scale=alt.Scale(range=["red","orange","lightblue"]))
)

background+cities
```

```
import folium
# lat y luego lon
# Por encima de los 100 marcadores, va lento.

m = folium.Map(location=[40.4, 3.6750],zoom_start=5)

temp = weather_df.groupby("city").mean().reset_index()

for i, row in temp.iterrows():
  marker = folium.Marker([row["lat"], row["lon"]], tooltip=row["temp"])
  marker.add_to(m)

m
```

```
m2 = folium.Map(location=[40.4, 3.6750],zoom_start=4, tiles = 'Stamen Toner')

temp = weather_df.groupby("city").mean().reset_index()

for i, row in temp.iloc[:1000].iterrows():
  marker = folium.Marker([row["lat"], row["lon"]], tooltip=row["temp"])
  marker.add_to(m2)

m2
```

```
select_city = alt.selection(type="interval",encodings=["x"])

countries = alt.topo_feature("https://cdn.jsdelivr.net/npm/world-atlas@2/countries-110m.json",
                             feature="countries")

background = alt.Chart(countries).mark_geoshape(
    fill="lightgrey",
    stroke="white"
)

temp = weather_df.groupby("city").mean().reset_index()

cities = alt.Chart(temp).mark_point().encode(
    latitude="lat",
    longitude="lon",
    size=alt.value(1),
    color=alt.Color("temp",scale=alt.Scale(range=["red","orange","lightblue"]))
).transform_filter(
    select_city
).properties(
    height=800,
    width=800
)

ranking_city = alt.Chart(temp).mark_bar().encode(
    x=alt.X("city",
            sort=alt.Sort(field="temp",
                          op="mean",
                          order="descending"),
            
            axis=None),
    y="mean(temp)",
    tooltip="city"
).properties(
    width=800,
    height=100,
    title="Ranking de ciudades por temperatura media",
    selection=select_city
)

ranking_city&(background+cities)
```
