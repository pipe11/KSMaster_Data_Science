# Funciones

Las definimos con `def` y devuelven un único valor.

##### Si una función no tiene un `return`, siempre devuelve `None`

```python
def distance(point1, point2):
    x=point2[0] - point1[0]
    y=point2[1] - point1[1]

    return (x**2 +y**2) **0.5
```

```python
distance
```

## Keyword arguments

```python
def distance(point1=(1,1), point2=(0,0)):
    
    x=point2[0] - point1[0]
    y=point2[1] - point1[1]

    return (x**2 +y**2) **0.5
```

```python
distance([3,0])
```

```python
x=6
print(id(x))
def multi(i):
    i.append()
    print(id(x))
    x=42 
    print(x, "Inside")
    
multi()
print(x, "Outside")
```

```python
x=6
def multi():
    global x
    x=42 
    print(x, "Inside")
    
multi()
print(x, "Outside")
```

```python
x=[1,2,3]
def multi(y):
    y.append(4) 
    print(y, "Inside")
    
multi(x.copy())
print(x, "Outside")
```

```python
x=[1,2,3]
print(id(x))
def multi():
    print(id(x))
    x.append(4) 
    print(x, "Inside")
    
multi()
print(x, "Outside")
```

## El keyword `pass`

No hace nada. Nos puede servir cuando Python requiere un bloque de código pero no estamos listos para escribir nada todavía.

```python
def this_is_function():
    pass
```

# Módulos

Archivos de texto con extensión `.py` que usamos para organizar nuestro código.

- import module/file (as mo)         (file no extension)
- from module/file import A (as B)
 - calling:    
   - module.func or 
   - mo.var  or 
   - A or  
   - B
- from module import *               (call original name)

```python
import math
```

```python
math.ceil(0.3)
```

```python
import math as m
```

```python
m.ceil(0.2)
```

```python
from math import *
```

```python
ceil(0.8)
```

```python
from math import ceil as c
```
